Example 1: Inserting nodes at the beginning and end of a doubly linked list.

````python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
        self.prev = None

class DoublyLinkedList:
    def __init__(self):
        self.head = None

    def insert_at_beginning(self, data):
        new_node = Node(data)
        if self.head is None:
            self.head = new_node
        else:
            new_node.next = self.head
            self.head.prev = new_node
            self.head = new_node

    def insert_at_end(self, data):
        new_node = Node(data)
        if self.head is None:
            self.head = new_node
        else:
            current = self.head
            while current.next:
                current = current.next
            current.next = new_node
            new_node.prev = current

````

Example 2: Deleting a node from a doubly linked list.

````python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
        self.prev = None

class DoublyLinkedList:
    def __init__(self):
        self.head = None

    def delete(self, data):
        current = self.head
        while current:
            if current.data == data:
                if current.prev:
                    current.prev.next = current.next
                else:
                    self.head = current.next
                if current.next:
                    current.next.prev = current.prev
                break
            current = current.next
````

Example 3: Reversing a doubly linked list.

````python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
        self.prev = None

class DoublyLinkedList:
    def __init__(self):
        self.head = None

    def reverse(self):
        current = self.head
        while current:
            temp = current.prev
            current.prev = current.next
            current.next = temp
            current = current.prev
        if temp:
            self.head = temp.prev
````

# 이중연결리스트 왼쪽, 오른쪽 순회
* 이중 연결 리스트(doubly linked list)는 각각의 노드가 이전 노드와 다음 노드의 포인터를 가지고 있는 자료 구조입니다. 각 노드는 데이터 필드(data field)와 두 개의 포인터(prev, next)로 구성됩니다. 이중 연결 리스트는 일반적인 연결 리스트와 달리, 노드를 양방향으로 탐색할 수 있어서 뒤에서부터도 탐색이 가능합니다.

* 이중 연결 리스트에서 가장 첫 번째 노드를 가리키는 포인터를 head라고 부릅니다. head는 리스트를 처음부터 순회하는 데 사용됩니다. head의 prev 포인터는 None을 가리키며, head의 next 포인터는 리스트에서 첫 번째 노드를 가리킵니다.

* 이중 연결 리스트를 순회할 때는 head부터 시작해서 각 노드의 next 포인터를 따라가면 됩니다. 이 때, 리스트를 끝까지 순회할 때까지 반복하면 됩니다. 만약 뒤에서부터 리스트를 순회하려면, 마지막 노드를 가리키는 포인터를 가지고 있어야 합니다. 이 경우, 마지막 노드를 가리키는 포인터를 tail이라고 부르고, 리스트를 끝까지 순회할 때는 tail부터 시작해서 각 노드의 prev 포인터를 따라가면 됩니다.


## 예제 1

````python
class Node:
    def __init__(self, data=None):
        self.data = data
        self.prev = None
        self.next = None

class DoublyLinkedList:
    def __init__(self):
        self.head = None
    
    def add_node(self, data): # 맨 뒤에 삽입
        new_node = Node(data)
        
        if self.head is None:
            self.head = new_node
        else:
            cur = self.head
            while cur.next:
                cur = cur.next
            cur.next = new_node
            new_node.prev = cur
````

이중 연결 리스트의 노드를 왼쪽에서 오른쪽으로 순회하는 코드와 오른쪽에서 왼쪽으로 순회하는 코드를 각각 작성해보겠습니다.

왼쪽에서 오른쪽으로 순회하는 코드는 다음과 같습니다.

````python
def traverse_left_to_right(self): # 순회하면서 출력
    cur = self.head
    while cur:
        print(cur.data)
        cur = cur.next
````

오른쪽에서 왼쪽으로 순회하는 코드

````python
def traverse_right_to_left(self):
    cur = None
    if self.head is not None:
        cur = self.head
        while cur.next is not None:
            cur = cur.next
    while cur is not None:
        print(cur.data)
        cur = cur.prev
````

각각의 코드에서는 현재 노드를 가리키는 포인터 cur를 사용하여 리스트를 순회하며, print 함수를 사용하여 각 노드의 데이터를 출력합니다. traverse_left_to_right 함수에서는 cur 포인터를 cur.next로 이동시키면서, 리스트를 왼쪽에서 오른쪽으로 순회합니다. traverse_right_to_left 함수에서는 cur 포인터를 리스트의 마지막 노드로 이동시킨 후, cur 포인터를 cur.prev로 이동시키면서, 리스트를 오른쪽에서 왼쪽으로 순회합니다.

# 예제2

self.head와 self.tail이 있는 코드는 이중 연결 리스트에서 노드를 추가, 삭제 및 탐색하는데 유용합니다. 이중 연결 리스트에서는 각 노드가 이전 노드와 다음 노드를 가리키는 포인터를 가지고 있기 때문에, 리스트의 처음과 끝을 가리키는 포인터를 유지하는 것이 중요합니다. 이를 위해 self.head와 self.tail을 사용할 수 있습니다.

아래는 self.head와 self.tail이 있는 이중 연결 리스트의 예시 코드입니다.

````python
class Node:
    def __init__(self, data=None):
        self.data = data
        self.prev = None
        self.next = None

class DoublyLinkedList:
    def __init__(self):
        self.head = None
        self.tail = None
    
    def add_node(self, data):
        new_node = Node(data)
        
        if self.head is None:
            self.head = new_node
            self.tail = new_node
        else:
            cur = self.tail
            cur.next = new_node
            new_node.prev = cur
            self.tail = new_node
            
    def remove_node(self, data):
        if self.head is None:
            return
        
        cur = self.head
        while cur:
            if cur.data == data:
                if cur == self.head and cur == self.tail:
                    self.head = None
                    self.tail = None
                elif cur == self.head:
                    self.head = cur.next
                    cur.next.prev = None
                elif cur == self.tail:
                    self.tail = cur.prev
                    cur.prev.next = None
                else:
                    cur.prev.next = cur.next
                    cur.next.prev = cur.prev
                return
            cur = cur.next
            
    def traverse_left_to_right(self):
        cur = self.head
        while cur:
            print(cur.data)
            cur = cur.next
    
    def traverse_right_to_left(self):
        cur = self.tail
        while cur:
            print(cur.data)
            cur = cur.prev
````

위 코드에서 add_node 함수는 리스트의 끝에 새로운 노드를 추가합니다. 노드를 추가할 때, 노드가 리스트에 처음으로 추가되는 경우 self.head와 self.tail이 새로운 노드를 가리키도록 하고, 그렇지 않은 경우 self.tail이 새로운 노드를 가리키도록 합니다.

remove_node 함수는 주어진 데이터를 가지는 노드를 찾아서 제거합니다. 제거할 노드가 리스트의 처음이나 끝인 경우 self.head와 self.tail을 갱신해줍니다.

traverse_left_to_right 함수와 traverse_right_to_left 함수는 각각 리스트를 왼쪽에서 오른쪽으로, 오른쪽에서 왼쪽으로 순회하면서 각 노드의 데이터를 출력합니다. 이 때, 각 함수에서는 각각 self.head와 self.tail을 시작점으로 사용하여 리스트를 순회합니다.
